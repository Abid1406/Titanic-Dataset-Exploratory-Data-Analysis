# Titanic Dataset - Exploratory Data Analysis (EDA)

## 1. Import Required Libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from data_profiling import ProfileReport

pd.set_option('display.max_columns', None)
```

## 2. Load Dataset

```
df = pd.read_csv(r"D:\internship\dataset\titanic.csv")
```

## 3. Display First 5 Rows

```
df.head()
```

## 4. Shape of Dataset

```
print("Rows and Columns:", df.shape)
```

## 5. Column Names

```
print(df.columns)
```

## 6. Dataset Information

```
df.info()
```

## 7. Statistical Summary - Numerical Columns

```
df.describe()
```

## 8. Statistical Summary - Categorical Columns

```
df.describe(include='object')
```

## 9. Check Missing Values

```
df.isnull().sum()
```

## 10. Missing Values Percentage

```
missing_percentage = (df.isnull().sum() / len(df)) * 100
missing_percentage
```

## 11. Visualize Missing Values

```
plt.figure(figsize=(10,6))

sns.heatmap(df.isnull(),
            cbar=False,
            cmap='viridis')

plt.title("Missing Values Heatmap")
plt.show()
```

## 12. Check Duplicate Values

```
df.duplicated().sum()
```

## 13. Data Cleaning

```
df['Age'] = df['Age'].fillna(df['Age'].median())

df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

df.drop('Cabin', axis=1, inplace=True)
```

## 14. Verify Missing Values Again

```
df.isnull().sum()
```

## 15. Survival Count

```
plt.figure(figsize=(6,4))

sns.countplot(x='Survived', data=df)

plt.title("Survival Count")
plt.show()
```

## 16. Gender Distribution

```
plt.figure(figsize=(6,4))

sns.countplot(x='Sex', data=df)

plt.title("Gender Distribution")
plt.show()
```

## 17. Passenger Class Distribution

```
plt.figure(figsize=(6,4))

sns.countplot(x='Pclass', data=df)

plt.title("Passenger Class Distribution")
plt.show()
```

## 18. Age Distribution

```
plt.figure(figsize=(8,5))

sns.histplot(df['Age'], bins=30, kde=True)

plt.title("Age Distribution")
plt.show()
```

## 19. Fare Distribution

```
plt.figure(figsize=(8,5))

sns.histplot(df['Fare'], bins=30, kde=True)

plt.title("Fare Distribution")
plt.show()
```

## 20. Survival Based on Gender

```
plt.figure(figsize=(6,4))

sns.countplot(x='Sex',
              hue='Survived',
              data=df)

plt.title("Survival based on Gender")
plt.show()
```

## 21. Survival Based on Passenger Class

```
plt.figure(figsize=(6,4))

sns.countplot(x='Pclass',
              hue='Survived',
              data=df)

plt.title("Survival based on Passenger Class")
plt.show()
```

## 22. Age vs Survival

```
plt.figure(figsize=(8,5))

sns.boxplot(x='Survived',
            y='Age',
            data=df)

plt.title("Age vs Survival")
plt.show()
```

## 23. Fare vs Survival

```
plt.figure(figsize=(8,5))

sns.boxplot(x='Survived',
            y='Fare',
            data=df)

plt.title("Fare vs Survival")
plt.show()
```

## 24. Convert Categorical Columns

```
df_encoded = df.copy()

df_encoded['Sex'] = df_encoded['Sex'].map({
    'male':0,
    'female':1
})

df_encoded['Embarked'] = df_encoded['Embarked'].map({
    'S':0,
    'C':1,
    'Q':2
})
```

## 25. Correlation Matrix

```
corr = df_encoded.corr(numeric_only=True)

plt.figure(figsize=(10,8))

sns.heatmap(corr,
            annot=True,
            cmap='coolwarm')

plt.title("Correlation Matrix")
plt.show()
```

## 26. Outlier Detection - Fare

```
plt.figure(figsize=(8,5))

sns.boxplot(x=df['Fare'])

plt.title("Fare Outliers")
plt.show()
```

## 27. Outlier Detection - Age

```
plt.figure(figsize=(8,5))

sns.boxplot(x=df['Age'])

plt.title("Age Outliers")
plt.show()
```

## 28. Pairplot

```
sns.pairplot(df[['Survived',
                 'Age',
                 'Fare',
                 'Pclass']])

plt.show()
```
